#calculation of gradients in TF

```python
x = tf.random.normal(shape=(2,2))
y = tf.random.normal(shape=(2,2))
```

$f(x,y) = \sqrt{(x^2 + y^2)}$

$\nabla f(x,y) = \frac{\partial f}
{\partial x} \hat{\imath} + \frac{\partial f}
{\partial y} \hat{\jmath}$


```python
#tf.gradienttape() will help in calculate gradients..

with tf.GradientTape() as tape:
  tape.watch(x)     ###to calculate gradient of x
  f = tf.sqrt(tf.square(x) + tf.square(y))

  df_dx = tape.gradient(f, x)

  print(df_dx)
```

```python
#tf.gradienttape() will help in calculate gradients..

with tf.GradientTape() as tape:
  tape.watch(y)     ###to calculate gradient of x & y
  tape.watch(x)
  f = tf.sqrt(tf.square(x) + tf.square(y))

  df_dx,df_dy = tape.gradient(f, [x,y]) #partial derivitives for both
  print(df_dx)
  print(df_dy)
```

#Learning to find simple linear regression
Example

$f(x) =  W.x+b$

```python
#creating a dummy data...
TRUE_W = 3.0
TRUE_B = 2.0

NUM_EXAMPLES = 1000

x = tf.random.normal(shape=[NUM_EXAMPLES])
```

```python
noise = tf.random.normal(shape=[NUM_EXAMPLES])
```

```python
y = x * TRUE_W + TRUE_B + noise
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, c="b")
plt.show()
```

#Lets define a model

```python
class Model(tf.Module):
  def __init__(self, **kwargs):
    super().__init__(**kwargs)

    #initial weights
    self.w = tf.Variable(5.0)
    self.b = tf.Variable(0.0)

  def __call__(self, x):
    return self.w*x + self.b
```

```python
model = Model()
```

```python
model(4)   # w = 5.0 = 4 * 5 = 20
```

```python
model.variables  #to see defined variables..
```

#Define the Loss Function

```python
 def MSE_loss(target_y, predicted_y):
  error =  target_y - predicted_y
  squared_error = tf.square(error)
  mse = tf.reduce_mean(squared_error)
  return mse
```

```python
plt.scatter(x,y, c="b")
#model without training...
pred_y = model(x)
plt.scatter(x, pred_y, c="r")

plt.show()
```

```python
current_loss = MSE_loss(y, model(x))
current_loss.numpy()
```

#Training def function code

```python
def train(model, x, y, learning_rate):

  with tf.GradientTape() as tape:
    current_loss = MSE_loss(y, model(x))


  dc_dw, dc_db = tape.gradient(current_loss, [model.w, model.b])

  model.w.assign_sub(learning_rate * dc_dw)
  model.b.assign_sub(learning_rate * dc_db)
```

```python
model = Model()

Ws, bs = [], []

epochs = 10*2

learning_rate = 0.1

w = model.w.numpy()
b = model.b.numpy()

init_loss = MSE_loss(y, model(x)).numpy()

print(f"Initial W: {w}, Initial bias: {b}, Initial loss: {init_loss}")
```

```python
for epoch in range(epochs):
  train(model, x, y, learning_rate)

  Ws.append(model.w.numpy())
  bs.append(model.b.numpy())

  current_loss = MSE_loss(y, model(x))

  print(f"for epoch: {epoch}, W: {Ws[-1]}, b: {bs[-1]}, current_loss: {current_loss}")
```

```python
plt.plot(range(epochs), Ws, "r", range(epochs), bs, "b")
plt.plot([TRUE_W] * epochs, "r--", [TRUE_B] * epochs, "b--")
plt.legend(["w", "b", "True W", "True B"])

plt.show()
```

```python
plt.scatter(x,y, c="b")
#model wit training...
pred_y = model(x)
plt.scatter(x, pred_y, c="r")

plt.show()
```
